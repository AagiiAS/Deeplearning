```python
import torch 
import matplotlib.pyplot as plt 
```

```python
#函数
x = torch.linspace(-20,20,1000)
# x
y = torch.sigmoid(x)
plt.plot(x,y)
plt.grid()
```

```python
#导函数
#requires_grad = compute grad 
x = torch.linspace(-20,20,1000, requires_grad=True)
## Sum the values to get a single scalar value
torch.sigmoid(x).sum().backward()
# x.detach = stopping gradient tracking
# x.grad = accessing gradient 
plt.plot(x.detach(), x.grad)
plt.grid()
```
